```python
import matplotlib.pyplot as plt
import pandas as pd

"""
n_stories: The choice of n depends on the user,
    n could be increase or decreased
"""

models_to_compare = [
    'TS-1M',
    'TS-3M',
    'TS-8M',
    'TS-28M',
    'TS-33M',
    'TS-1Layer-21M',
    'TinyLLaMa',
    'GPT2-small'
]

n_stories = 5

for model in models_to_compare:
    !python3 three/gpt4_eval.py -m "{model}" -n 5 -d cpu -s "output/graph_data/{model}_{n_stories}_results.csv"
```

```python
test_results = {
    'grammar': [],
    'creativity': [],
    'consistency': []
}

for model in models_to_compare:
    result_name = f"output/graph_data/{model}_{n_stories}_results.csv"
    results_df = pd.read_csv(result_name)
    metrics = results_df[['grammar', 'creativity', 'consistency']]
    metrics_avg = metrics.mean(axis=0)
    metrics_avg = metrics_avg.to_dict()
    for metric in metrics_avg:
        test_results[metric].append(metrics_avg[metric])

test_results
```

```python
df = pd.DataFrame(test_results, index=models_to_compare)

df.T.plot(kind='line', marker='o', figsize=(14, 10))
plt.title('Performance of LLMs on Different Tests')
plt.xlabel('Tests')
plt.ylabel('Performance')
plt.legend(title='LLMs')
plt.show()
```
